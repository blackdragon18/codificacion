## Draw the computation graph for binary classification with hinge loss

In Lecture 3, we learned the computation graph for the hinge loss function below. We can use Torchviz to plot this graph as follows (upside down):

Screenshot 2025-02-17 at 11.09.25 AM.png

```python
# This code is partially generated using ChatGPT for demonstration purposes.
# Reviewed and modified to fit the lecture content.

import torch
from torchviz import make_dot

# Define input (x) and label (y)
x = torch.tensor(2.0)  # Input feature
y = torch.tensor(1.0)  # Ground truth label (+1 or -1)

# Define parameter (w), requires gradient since it's trainable
w = torch.tensor(1.0, requires_grad=True)

# Compute score
score = w * x * y  # Linear transformation

# Compute hinge loss: max(1 - score, 0)
margin = 1 - score
loss = torch.max(margin, 0)  # Equivalent to max(1 - score, 0)

# Generate computational graph
dot = make_dot(loss, params={'w': w})
dot.render("hinge_loss_graph", format="png")  # Save graph
dot  # Display graph
```

```python
# This code is partially generated using ChatGPT for demonstration purposes.
# Reviewed and modified to fit the lecture content.

import torch
from torchviz import make_dot

# Define input (x) and label (y)
x1 = torch.tensor(2.0)  # Input feature
x2 = torch.tensor(3.0)  # Input feature
#x3 = torch.tensor(4.0)  # Input feature

y = torch.tensor(1.0)  # Ground truth label (+1 or -1)

# Define parameter (w), requires gradient since it's trainable
v1 = torch.tensor(1.0, requires_grad=True)
v2 = torch.tensor(1.0, requires_grad=True)
#v3 = torch.tensor(1.0, requires_grad=True)

w1 = torch.tensor(1.0, requires_grad=True)
w2 = torch.tensor(1.0, requires_grad=True)
#w3 = torch.tensor(1.0, requires_grad=True)

# Compute score
loss = (w1 * torch.sigmoid(v1 * x1) + w2 * torch.sigmoid(v2 * x2) - y) ** 2

# Generate computational graph
dot = make_dot(loss, params={'w': w})
dot.render("hinge_loss_graph", format="png")  # Save graph
dot  # Display graph
```
