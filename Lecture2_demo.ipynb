```python
import sys,re
sys.path.append('/content/cs541/lectures/sentiment/')

import submission, util
#from autumn2019.assignments.sentiment import util
from collections import defaultdict

# Read in examples
trainExamples = util.readExamples('/content/cs541/lectures/lecture3/email2.train')
devExamples = util.readExamples('/content/cs541/lectures/lecture3/email2.dev')

def max_consecutive_sequence(password, pattern):
    sequences = re.findall(pattern, password)
    return max(map(len, sequences), default=0)

def max_repeated_sequence(password):
    max_count = 1
    current_count = 1

    for i in range(1, len(password)):
        if password[i] == password[i-1]:
            current_count += 1
            max_count = max(max_count, current_count)
        else:
            current_count = 1

    return max_count

def featureExtractor(x):  # phi(x)
    # x = "took Mauritius into"
    phi = defaultdict(float)
    phi["length"] = len(x)
    length = phi["length"]
    num_letters = sum(c.isalpha() for c in x)
    num_digits = sum(c.isdigit() for c in x)
    num_special = sum(not c.isalnum() for c in x)



    phi["letter_ratio"] = num_letters / length if length > 0 else 0
    phi["digit_ratio"] = num_digits / length if length > 0 else 0
    phi["special_ratio"] = num_special / length if length > 0 else 0

    phi["max_consecutive_special"] = max_consecutive_sequence(x, r'[!-@]+')


    # switch_count = 0
    # for i in range(1, len(x)):
    #     if (not x[i-1].isalnum() and x[i].isalnum()) or (not x[i].isalnum() and x[i-1].isalnum()):
    #         switch_count += 1
    # phi["switch_count"] = switch_count

    phi["max_consecutive_alnum"] = max_consecutive_sequence(x, r'[A-Za-z0-9]+')
    phi["max_consecutive_upperletter"] = max_consecutive_sequence(x, r'[A-Za-z]+')
    #phi["max_consecutive_lowerletter"] =  max_consecutive_sequence(x, r'[a-z]+')
    #phi["max_consecutive_aeiou"] =  max_consecutive_sequence(x, r'[aeiou]+')
    phi["max_consecutive_number"] = max_consecutive_sequence(x, r'[0-9]+')
    #print(x, phi)
    return phi

# Learn a predictor
weights = submission.learnPredictor(trainExamples, devExamples, featureExtractor, 100, 0.0001)
util.outputWeights(weights, 'weights')
util.outputErrorAnalysis(devExamples, featureExtractor, weights, 'error-analysis')

# Test!!!
testExamples = util.readExamples('/content/cs541/lectures/lecture3/email2.test')
predictor = lambda x : 1 if util.dotProduct(featureExtractor(x), weights) > 0 else -1
print('test error =', util.evaluatePredictor(testExamples, predictor))


```
